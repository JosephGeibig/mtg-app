```python
import requests
import pandas as pd
import json

#Gets a list of all commander-legal sets 
def getsets():
    url = "https://api.scryfall.com/sets"
    response = requests.get(url)
    correcttypes = ["expansion", "commander", "draft_innovation", "core", "from_the_vault", "masters","duel_deck","planechase","starter","box"]
    if response.status_code == 200:
        data = response.json()
        sets = [set_data["code"] for set_data in data["data"] if set_data["set_type"] in correcttypes]
        return sets
    else:
        print("failed")
        return
    
#Pulls a list of sets and reverses the order (so oldest sets come first in the list )
setlist = getsets()
setlist.reverse()
removelist = []
for i in setlist:
    if len(i) > 3:
        removelist.append(i)
for i in removelist:
    setlist.remove(i)
setlist.remove("q07")
print(setlist)
print(len(setlist))
```

```python
testval = 256
test1 = setlist[testval]
fullcardlist = []
allprevtest = ["Air Elemental", "Animate Artifact", "Abbey Griffin"]
def getcards(setcode, allprev):
    url = f"https://api.scryfall.com/cards/search?q=set%3A{setcode}+legal%3Acommander"

    listofsets = []
    namelist = []
    textlist = []
    cmclist = []
    ctypelist = []
    mvlist = []
    keywordlist = []
    colorslist = []

    data = []

    response = requests.get(url)
    if response.status_code == 200:
        pulled = response.json()
        
        # print(pulled["data"])
        data.append(pulled["data"])
        
        while pulled["has_more"] == True:
            nextpage = requests.get(pulled["next_page"]).json()
            data.append(nextpage["data"])
            pulled = nextpage
        
        listofcards = []

        if len(data) < 2:
            listofcards = data[0]

        if len(data) == 2:
            listofcards = data[0] + data[1]

        if len(data) == 3:
            listofcards = data[0] + data[1] + data[2]
            
        if len(data) == 4:
            listofcards = data[0] + data[1] + data[2] + data[3]

        if len(data) == 5:
            listofcards = data[0] + data[1] + data[2] + data[3] + data[4]

        if len(data) == 6:
            listofcards = data[0] + data[1] + data[2] + data[3] + data[4] + data[5]

        # print(data)


        for i in listofcards:
            if i["name"] not in allprev:
                    listofsets.append(setcode)
                    namelist.append(i["name"])

                    if i.get("oracle_text") is None: #splits it so that split cards can actually be fully read
                        face1 = i["card_faces"][0]
                        face2 = i["card_faces"][1]
                        oracle1 = face1["oracle_text"]
                        oracle2 = face2["oracle_text"]
                        textlist.append(oracle1 + "//" + oracle2)
                    else:
                        oracle = i["oracle_text"]
                        textlist.append(oracle)
                    
                    cmclist.append(i["cmc"])
                    ctypelist.append(i["type_line"])

                    if i.get("mana_cost") is None: #splits it so that split cards can actually be fully read
                        face1 = i["card_faces"][0]
                        face2 = i["card_faces"][1]
                        mv1 = face1["mana_cost"]
                        mv2 = face2["mana_cost"]
                        mvlist.append(mv1 + "//" + mv2)
                    else: 
                        mvlist.append(i["mana_cost"])

                    #Keywords needs to be changed coming up
                    keywordlist.append(i["keywords"])

                    if i.get("colors") is None: #splits it so that split cards can actually be fully read
                        face1 = i["card_faces"][0]
                        face2 = i["card_faces"][1]

                        colors1 = "".join(face1["colors"]) #concatenates the entire list
                        if colors1 == "":
                            colors1 = "Colorless"

                        colors2 = "".join(face2["colors"]) #concatenates the entire list
                        if colors2 == "":
                            colors2 = "Colorless"

                        colorslist.append(colors1 + "//" + colors2)
                    else: 
                        colors = "".join(i["colors"])
                        if colors == "":
                            colors = "Colorless"
                        colorslist.append(colors)             

        reslist = [list(item) for item in zip(listofsets, namelist, textlist, cmclist, ctypelist, mvlist, keywordlist, colorslist)]
        # print(reslist)
        # print(len(namelist), namelist)
    return reslist, namelist

testres, testname = getcards(test1, allprevtest)
testres
```

```python
allprev = []
cardinfolist = []
# count = 0
for i in setlist:
    print(i)
    # count += 1
    results, namelist = getcards(i,allprev)
    cardinfolist.extend(results)
    # if count == 10:
    #     break
    allprev.extend(namelist)
print("Check if duplicates were removed from list")
print(len(allprev) == len(set(allprev)))
print("Check if there are no duplicates in real cardlist")
print(len(allprev) == len(cardinfolist))
```

```python
#Parses our list of all cards into reasonable cards 
sets,name,text,cmc,ctype,mv,keywords,colors = [],[],[],[],[],[],[],[]
for i in cardinfolist:
    sets.append(i[0])
    name.append(i[1])
    text.append(i[2])
    cmc.append(i[3])
    ctype.append(i[4])
    mv.append(i[5])  
    keywords.append(i[6])
    colors.append(i[7])
```

```python
#Builds the final database based on the card info which was pulled prior
ldf = pd.DataFrame()
# ldf["setname"] = setname
ldf["firstset"] = sets
ldf["cardname"] = name
ldf["text"] = text
ldf["cmc"] = cmc
ldf["mana"] = mv
ldf["type"] = ctype
ldf["keywords"] = keywords
ldf["colors"] = colors

ldf
```
